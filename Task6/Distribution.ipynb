```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#1. Normal Paylanma
normal = np.random.normal(loc=50, scale=10, size=1000)
plt.hist(normal, bins=30,color='orange',edgecolor='white')
plt.title("Normal Paylanma")
plt.xlabel("Qiymet")
plt.ylabel("Tezlik")
plt.show()
```

```python
# 2.Uniform Paylanma
uniform = np.random.uniform(low=0, high=100, size=1000)
plt.hist(uniform, bins=30 , color='red',edgecolor='white')
plt.title("Uniform Paylanma")
plt.xlabel("Qiymet")
plt.ylabel("Tezlik")
plt.show()
```

```python
# 3.Poisson Paylanma
poisson = np.random.poisson(lam=5,size=1000)
plt.hist(poisson, bins=15,color='green',edgecolor='white')
plt.title("Poisson Paylanma")
plt.xlabel("Qiymet")
plt.ylabel("Tezlik")
plt.show()
```

```python
# 4.Right skewed
right_skewed=np.random.exponential(scale=2, size=1000)
plt.hist(right_skewed, bins=30,color='purple',edgecolor='white')
plt.title("Right Skewed Paylanma")
plt.xlabel("Qiymet")
plt.ylabel("Tezlik")
plt.show()
```

```python
# 5.Left skewed
left_skewed=-np.random.exponential(scale=2, size=1000)
plt.hist(left_skewed, bins=30,color='purple',edgecolor='white')

plt.title("Left Skewed Paylanma")
plt.xlabel("Qiymet")
plt.ylabel("Tezlik")
plt.show()
```
