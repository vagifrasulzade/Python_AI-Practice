```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#1. Normal Paylanma
np.random.seed(42)

normal = np.random.normal(loc=50, scale=10, size=1000)

plt.hist(normal, bins=30)
plt.title("Normal Paylanma")
plt.xlabel("Qiymet")
plt.ylabel("Tezlik")
plt.show()
```

```python
# 2.Uniform Paylanma
np.random.seed(42)
uniform = np.random.uniform(low=0, high=100, size=1000)
plt.hist(uniform, bins=30)
plt.title("Uniform Paylanma")
plt.xlabel("Qiymet")
plt.ylabel("Tezlik")
plt.show()
```

```python
# 3.Eksponensial Paylanma
np.random.seed(42)
eksponensial = np.random.exponential(scale=10,size=1000)
plt.hist(eksponensial, bins=30)
plt.title("Eksponensial Paylanma")
plt.xlabel("Qiymet")
plt.ylabel("Tezlik")
plt.show()
```

```python
# 4.Binomial Paylanma
np.random.seed(42)
binomial=np.random.binomial(n=10, p=0.5, size=1000)
plt.hist(binomial, bins=11)
plt.title("Binomial Paylanma")
plt.xlabel("Qiymet")
plt.ylabel("Tezlik")
plt.show()
```
